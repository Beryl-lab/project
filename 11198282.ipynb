```python
dataset.plot()
plt.title('A SCATTERED GRAPH OF THE DATASET')
plt.show()
```

```python
dataset.plot.scatter(x = 'quality' , y = 'volatile acidity')
plt.title('A SCATTERED GRAPH OF VOLATILE ACIDITY AGAINST QUALITY')
plt.xlabel('quality')
plt.ylabel('Alcohol')
plt.show()
```

```python
dataset.plot.scatter(x='quality' , y= 'alcohol')
plt.title(' A SCATTERED PLOT OF ALCOHOL AGAINST QUALITY')
plt.xlabel('Quality')
plt.ylabel('Alcohol')
plt.show
```

### **Reports On Plots**
<hr>

- The plot shows that as the quality increases, the volatile acidity decreases. The plot corresponds to the negative correlation found in the corrplots with volatile acidity and quality
- It is seen that the volatile acidity changes the most during the data analysis and hence, tells a lot about the wine.
- For volatile acids, the pH increases with acidity
- The factors which affected the quality of wine the most were the Alcohol percentage, Sulphate and Acid concentrations
- Density does not play a part in improving the wine quality
